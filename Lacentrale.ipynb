# What is Web Scraping ?

## Presentation of Web Scraping

*Web scraping* is the practice of using programming tools for gathering data from the internet. In order to realize a web scraping, we must analyze the structure of the website such as the HTML code, program a script that will allow us to collect the data necessary for our work and finally organize the data (see Figure 1). *Web scraping* is not a new term, in years past the practice has been more commonly known as *screen scraping*, *web harvesting* or similar variations.

These * web scraping* script are often developed using such programming languages as R and Python. This has
to do with the popularity of these languages in Data Science and availability libraries (e.g. rvest
package in R or Beautiful Soup library in Python) for automatic crawling and parsing of Web data. 

We can find on the internet a lot of useful information for our business but not properly presented. *Web  scraping* as several benefits. It permit to us to have the correct information to make the right decision. For example, we can scrape seloger.fr to compare the price of a house in different city to make a good investment decision or we can still use the bloomberg API to retrieve stock price data to build a predictive model. We also use websites based on webscraping such as flight or insurance comparison websites.

## The Legalities and Ethics of Web Scraping  

If Web Scraping overloads or damages a website or a Web server, then the person responsible for the
damage can be prosecuted under the trespass to chattels law (Dryer and Stockton 2013). However, the
damage needs to be material and easy to prove in court in order for the owner of the Web server to be
eligible for a financial compensation. 

Computer servers are costly, when a user browses the site it consumes the resources from the server. In the case of a web-scraping, we want to collect the data without damaging the server, we do not wish to launch a DDOS attack.
The risk is also to be blacklisted from the website.

Three criteria need to be meet for a web scraper to violate trespass to chattels :
- *Lack of consent* : Many websites Terms of Service agreements specifically prohibit the use of scrapers.
- *Actual harm* : In addition of servers, a scraper can also take a website down. 
- *Intentionality* : the author of the web-scraping is responsible.

During the first execution of the loop in order to retrieve our variables, we realized that lacentrale.fr sent us bad information concerning the price, the mileage and the year. Indeed, it is a technique set up by the site to block scrapers.

# Machine Learning to predict cars's price

After the data analysis, we decide to delete the outlier :

```python
df_final3 = df_final2[ 
        (df_final2.Mileage <= 200000) #Too many kilometers
      & (df_final2.annee >= 1999) #Too old
      & (df_final2.Price <= 200000) #Too expensive
      ]
```

In addition, we achieve our price prediction on the twenty most popular brands :

```python
import sklearn
X = df_final3[[ 'annee', 'Mileage', 'departement', 'Top_condition_car', 'TOP_GUARANTY', 'Model_version_length', 
                'TOP_SELLER', 'TOP_KIA', 'TOP_HYUNDAI', 'TOP_SMART', 'TOP_AUDI', 'TOP_FORD','TOP_SUZUKI', 'TOP_MAZDA', 
                'TOP_SKODA', 'TOP_SEAT', 'TOP_VOLVO', 'TOP_CITROEN', 'TOP_MINI', 'TOP_OPEL','TOP_PEUGEOT','TOP_TOYOTA', 
                'TOP_MERCEDES', 'TOP_NISSAN', 'TOP_RENAULT', 'TOP_BMW', 'TOP_VOLKSWAGEN']]
Y = df_final3['Price'] 
```

The train_test_split function retrieves 80% of the available rows to form the learning dataset. The remaining 20% is the test data set.
Before dividing the dataset into two, the function uses a pseudo random number generator to make sure that the test data set will contain information from all brands.

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .20, random_state = 42)
```

## Linear Regression

<div align="left"><span style="font-family:Calibri;font-size:25px;color:black"> <strong> Lasso
</strong></span></div>

Lasso, *Least Absolute Shrinkage and Selection Operator*, is a linear regression method. This method forces the coefficients closer to 0. So when we use this regression some coefficients are exactly 0, some variables are ignored in our model. It is an automatic selection of features. With some null coefficients, the model is simpler to interpret and reveals the most important variables of our model.

```python
from sklearn.linear_model import Lasso 
lasso = Lasso().fit(X_train, Y_train)
```

```python
print('Training set score: {:.2f}'.format(lasso.score(X_train, Y_train)))
print('Test set score: {:.2f}'.format(lasso.score(X_test, Y_test)))
print("Number of features used : {}".format(np.sum(lasso.coef_ !=0)))
```

As we can see, lasso gives weak results, both on the test data set and on the learning data set. This indicates an underfitting, and we notice that all the variables are used during the regression.
The lasso regression has a regularization parameter, *aplha*, which controls the impact of the method on the value of the coefficients. By default, the *alpha* parameter is 1. To try to set the underfitting, we will decrease the alpha parameter and the value of the maximum number of iterations.

```python
lasso001 = Lasso(alpha=0.01, max_iter=100000).fit(X_train, Y_train)
print('Training set score: {:.2f}'.format(lasso001.score(X_train, Y_train)))
print('Test set score: {:.2f}'.format(lasso001.score(X_test, Y_test)))
print("Number of features used : {}".format(np.sum(lasso001.coef_ !=0)))
```

However, in our case, we can not solve the problem of underfitting by modifying the alpha parameter. We will try to add new variables to our model so that it can learn better. 

We will add:
- Mileage^2
- Mileage^3
- Model_version_length^2
- Model_version_length^3
- TOP_FRANCE : if it's a french mark
- TOP_GERMANY : if it's a german mark
- We replace the variable department by the dummy TOP_IDF : If the ad islocated in Paris

```python
df_final3['Mileage^2']= df_final3['Mileage']**2
df_final3['Model_version_length^2']= df_final3['Model_version_length']**3
df_final3['Mileage^3']= df_final3['Mileage']**2
df_final3['Model_version_length^3']= df_final3['Model_version_length']**3
```

```python
df_final3["TOP_FRANCE"] = np.where((df_final3['TOP_RENAULT'] == 1) | (df_final3["TOP_PEUGEOT"] == 1) | (df_final3["TOP_CITROEN"] == 1), 1,0)
df_final3["TOP_GERMANY"] = np.where((df_final3['TOP_MERCEDES'] == 1) | (df_final3["TOP_AUDI"] == 1) | (df_final3["TOP_BMW"] == 1) | (df_final3["TOP_VOLKSWAGEN"] == 1) | (df_final3["TOP_SMART"] == 1)| (df_final3["TOP_OPEL"] == 1), 1,0)
```

```python
df_final3["TOP_IDF"] = np.where((df_final3['departement'] == 75) | (df_final3['departement'] == 77) | (df_final3['departement'] == 78)
                         | (df_final3['departement'] == 91)| (df_final3['departement'] == 92)| (df_final3['departement'] == 93)
                         | (df_final3['departement'] == 94)| (df_final3['departement'] == 95), 1,0)
```

We start again the Lasso method hoping to have enough regressors.

```python
X = df_final3[[ 'annee', 'Mileage','Mileage^2','Mileage^3', 'Top_condition_car', 'TOP_IDF', 'TOP_GUARANTY', 'Model_version_length',
         'Model_version_length^2', 'Model_version_length^3','TOP_SELLER', 'TOP_KIA', 'TOP_HYUNDAI', 
         'TOP_SMART', 'TOP_AUDI', 'TOP_FORD','TOP_SUZUKI', 'TOP_MAZDA', 'TOP_SKODA', 'TOP_SEAT', 'TOP_VOLVO', 'TOP_CITROEN', 
         'TOP_MINI', 'TOP_OPEL','TOP_PEUGEOT','TOP_TOYOTA', 'TOP_MERCEDES', 'TOP_NISSAN', 'TOP_RENAULT', 'TOP_BMW', 
         'TOP_VOLKSWAGEN', 'TOP_FRANCE','TOP_GERMANY']]
Y = df_final3['Price'] 
```

```python
lasso001 = Lasso(alpha=0.01, max_iter=100000).fit(X_train, Y_train)
print('Training set score: {:.2f}'.format(lasso001.score(X_train, Y_train)))
print('Test set score: {:.2f}'.format(lasso001.score(X_test, Y_test)))
print("Number of features used : {}".format(np.sum(lasso001.coef_ !=0)))
```

As we can see, even adding variables the result of the Lasso has not changed. In the case of linear regression, we do not have enough regressors to accurately predict the price of a car.

## Ensemble Methods

<div align="left"><span style="font-family:Calibri;font-size:25px;color:black"> <strong> Gradient Boosting
</strong></span></div>

Each tree built by the boosting gradient tries to correct the errors made by the previous one. The trees created are not deep. 

The maximum depth parameter limits the number of nodes in the tree. And the learning_rate parameter controls the degree of intensity with which each tree tries to correct the errors of previous trees.

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

gbr = GradientBoostingRegressor(loss ='ls', learning_rate=0.5, max_depth=6)
gbr.fit (X_train, Y_train)
predicted = gbr.predict(X_test)
residual = Y_test - predicted

fig = plt.figure(figsize=(30,30))
ax1 = plt.subplot(211)
sns.distplot(residual, color ='teal')
plt.tick_params(axis='both', which='major', labelsize=20)
plt.title('Residual counts',fontsize=35)
plt.xlabel('Residual',fontsize=25)
plt.ylabel('Count',fontsize=25)

plt.show()

rmse = np.sqrt(mean_squared_error(Y_test, predicted))
print(rmse)
```

```python
print("Accuracy on training set: {:.3f}".format(gbr.score(X_train, Y_train))) 
print("Accuracy on test set: {:.3f}".format(gbr.score(X_test, Y_test))) 
```

We get better results with the *gradiant boosting*.
However, as we have seen on the data analysis section, the average price of a car is 23985. This is why a prediction error of 12583  is not satisfactory.
This may mean that our variables are not explanatory enough to provide good prediction, or that the model is not powerful enough.
